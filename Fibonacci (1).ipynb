```python
import matplotlib.pyplot as plt
import numpy as np
from time import time

def fibo(n):
    i = 0
    j = 1
    k = 0
    while k < n :
        j = i + j
        i = j - i
        k = k + 1
    return i;
    
a = []
b = []
for j in range(1000):
    for i in range(1000):
        t0 = 1000000000*time()
        fibo(j)
        t1 = 1000000000*time() - t0
        a.append(t1)
    a = np.array(a)
    b.append(np.mean(a))
    a = []

b = np.array(b)
print(b)
```

```python
plt.scatter(range(1000), b)
plt.xlabel('n')
plt.ylabel('time (ns)')
plt.title('Tiempo de ejecucion');
```
